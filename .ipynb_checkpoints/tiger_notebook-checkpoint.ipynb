# Tiger analyze

```python
import pymongo
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from bson.code import Code
```

### Imports

Importing our libraries. We will use matplotlib to plot our wordcloud.<br>
Using Code from bson to map reduce and count words.

```python
mongo_connection = pymongo.MongoClient("localhost", 27017)
documents = mongo_connection.tiger.documents
# Loading our whole database to a list
bundle_list = list(
    documents.aggregate([
            {"$match" : {"yandexTranslationBody" : {"$exists" : True}, "$and": [ { "yandexTranslationBody": { "$ne": "" } } ]}}
        ],allowDiskUse=True)
)
title_list = list(
    documents.aggregate([
            {"$match" : {"yandexTranslationTitle" : {"$exists" : True}, "$and": [ { "yandexTranslationTitle": { "$ne": "" } } ]}}
        ], allowDiskUse=True)
)
```

### Initialization

Creating a connection to our mongodb and getting our collections into lists. We only need the documents which have a yandex translation. <br>
The second list contains non-null translated titles.

```python
# Let's get the string for our titles and texts
str_text = ""
str_title = ""
nb_containing_bag = 0
for doc in bundle_list:
    str_text += doc["yandexTranslationBody"]
    str_title += doc["yandexTranslationTitle"]
```

### String extraction

Getting a raw string in order to make a wordcloud. Check wordcloud documentation in order to get other shapes.

```python
# Wordcloud for our documents

wordcloud = WordCloud().generate(str_text)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
# Wordcloud for our titles
words = WordCloud().generate(str_title)
plt.imshow(words)
plt.axis("off")
plt.show()
```

### Wordclouds

Creating wordclouds for documents and titles. <br>
We can notice that the titles' wordcloud contains more interesting words that the document's one. Let's check if our documents' structure can help us classify them. <br>
We are going to compare the words appearances for both titles and documents' bodies and check if there is some "garbage".

```python
# Getting the most used words in the document bodies
map = Code(
"""
function() {  
    var summary = this.yandexTranslationBody;
    if (summary) { 
        // quick lowercase to normalize per your requirements
        summary = summary.toLowerCase().replace(/[.!–-]/g, " ");
        summary = summary.replace(/[(-);'",.?:/‘’]/g, " ");
        summary = summary.split(" ");
        for (var i = summary.length - 1; i >= 0; i--) {
            // might want to remove punctuation, etc. here
            if (summary[i].length < 50 && summary[i].length > 0)  {      // make sure there's something
               emit(summary[i], 1); // store a 1 for each word
            }
        }
    }
};
"""
)

reduce = Code(
"""
function( key, values ) {    
    var count = 0;    
    values.forEach(function(v) {            
        count +=v;    
    });
    return count;
}
"""
)

result = documents.map_reduce(map, reduce, "myresults")
```

```python
# Getting the most used words in the document titles
map_title = Code(
"""
function() {  
    var summary = this.yandexTranslationTitle;
    if (summary) { 
        // quick lowercase to normalize per your requirements
        summary = summary.toLowerCase().replace(/[.!–-]/g, " ");
        summary = summary.replace(/[(-);'",.?:/‘’|”]/g, " ");
        summary = summary.split(" ");
        for (var i = summary.length - 1; i >= 0; i--) {
            // might want to remove punctuation, etc. here
            if (summary[i].length < 50 && summary[i].length > 0)  {      // make sure there's something
               emit(summary[i], 1); // store a 1 for each word
            }
        }
    }
};
"""
)

reduce_title = Code(
"""
function( key, values ) {    
    var count = 0;    
    values.forEach(function(v) {            
        count +=v;    
    });
    return count;
}
"""
)

result_title = documents.map_reduce(map_title, reduce_title, "myresults2")
```

### Map-Reduce

Using mongodb to count the words in our documents and titles. Storing the results in temp collections. <br><br>
Facing a problem when map-reducing : "alive." and "alive" are considered as different words leading in counting errors.<br>
Workaround : getting rid of punctuation. We decided to replace ".", ",", "!", "?", "-", "'", '"', "(", ")" by a single space, so that the punctuation doesn't fake our count.

```python
# Getting the 100 most used words in our documents and their count
docfinal = list(result.find())
docsorted = sorted(docfinal, key=lambda x: x['value'], reverse=True)
docfinal_title = list(result_title.find())
docsorted_title = sorted(docfinal_title, key=lambda x: x['value'], reverse=True)
print("We actually got {} documents with a Yandex Translation and {} different words".format(len(bundle_list), len(docsorted)))
print("We also got {} different words in our {} titles\n".format(len(docsorted_title), len(title_list)))
print("\tKEY\t\tVALUE\t\tAVG (WORDS/DOC)\t\t\tKEY_TITLE\tVALUE\tAVG (WORDS/TITLE)\n")
# Formatting the result
for index, words, title in zip(range(1, 100), docsorted[0:100], docsorted_title[0:100]):
    if len(words["_id"]) < 8 and len(title["_id"]) < 8:
        print("{}\t{}\t\t{}\t\t{}\t\t{}\t\t{}\t{}".format(
                index, 
                words["_id"], 
                int(words["value"]), 
                words["value"]/len(bundle_list),
                title["_id"],
                int(title["value"]),
                title["value"]/len(title_list)
            ))
    elif len(words["_id"]) < 8 and len(title["_id"]) >= 8:
        print("{}\t{}\t\t{}\t\t{}\t\t{}\t{}\t{}".format(
                index, 
                words["_id"], 
                int(words["value"]), 
                words["value"]/len(bundle_list),
                title["_id"],
                int(title["value"]),
                title["value"]/len(title_list)
            ))
    elif len(words["_id"]) >= 8 and len(title["_id"]) < 8:
        print("{}\t{}\t{}\t\t{}\t\t{}\t\t{}\t{}".format(
                index, 
                words["_id"], 
                int(words["value"]), 
                words["value"]/len(bundle_list),
                title["_id"],
                int(title["value"]),
                title["value"]/len(title_list)
            ))
    else:
        print("{}\t{}\t{}\t\t{}\t\t{}\t{}\t{}".format(
                index, 
                words["_id"], 
                int(words["value"]), 
                words["value"]/len(bundle_list),
                title["_id"],
                int(title["value"]),
                title["value"]/len(title_list)
            ))
```

# Interpretation

#### Titles vs full documents

As we can see on the tab, the first non-standard word comes in 26th position for the documents. With the title, it comes in 4th position. We can also notice that there are more specific words on titles that on full documents : does this make the classification easier ?<br><br>
The average word count on titles shows that the specific ones we are looking for are not very common : < 0.10 words/title. We can imagine that those specific words will be highly valued by our neural network -> we need to check this by making a first classification and verifying if interesting titles contains those specifc words. If not, we will have to use complete documents, which is way more complicated.<br>
In other terms, we need to check if there is a correlation between a document's relevance and the words contained in its title. If there's a strong one, we may be able to classify it efficiently. If not, we will have to use whole documents.<br>
Above results shows that we should rether focus on titles than on whole documents : there seems to be a unifom words distribution on titles, which is not the case for bodies.
